```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(1,6)
b = a**2 - np.sqrt(a)
a_new = ['A','B','C','D','E']
plt.xticks(a, a_new, rotation = 0)
plt.scatter(a,b)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(1,6)
b = a**2 - np.sqrt(a)
a_new = ['A','B','C','D','E']
fig,ax = plt.subplots()
ax.set_xticks(a)
ax.set_xticklabels(a_new)
ax.scatter(a,b)
plt.show()
```
