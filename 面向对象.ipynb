```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import random

mpl.rcParams['font.family'] = 'STFangsong'
mpl.rcParams['font.weight'] = 'medium'
mpl.rcParams['grid.linestyle'] = '-.'
mpl.rcParams['lines.linewidth'] = 1

col = ['r','g','k','y','b','0.5']
lsty = ['-','--','-.']
width = [x for x in range(1,4)]
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)
x = np.linspace(0,10,100)
ax = [ax1,ax2,ax3,ax4,ax5,ax6]
ylist = [np.sin(x),np.cos(x),np.tan(x),np.sqrt(x),np.sin(x) + np.cos(x),np.tan(x) - np.sqrt(x)]
for i in range(0,6):
    ax[i].plot(x,ylist[i],c = random.choice(col),ls = random.choice(lsty),lw = random.choice(width))

plt.show()
```
