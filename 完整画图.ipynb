```python
#本程序基于Python3.6进行的，运行的是Anaconda中的Jupyter Notebook，基于网页运行的。
#以下为导入画图的Python库
import xlrd
import numpy as np
import matplotlib.pyplot as plt

#加载外部数据，数据为txt格式，有两列组成
data = np.loadtxt('C:/Users/12/Desktop/ts.txt')

x = data[:,0] #取出数据的第一列给x
y = data[:,1] #取出数据的第二列给y
fig,ax = plt.subplots() #进行面向对象操作，固定格式
ax.plot(x,y) #画图
ax.set_title('Relationship between moment and time') #给图加图名
fig.set_size_inches(16,9) #设置图片的尺寸大小

plt.plot(x,y, ls = '--',c = 'g', label = 'momemt') #设置曲线的格式，包括线型，颜色和标签
plt.legend() #显示图例
plt.grid(True, lw = 0.5, ls = '--', color = '0') #刻度网格的显示，如果不喜欢将True改成False

plt.xlabel('Time [s]') #x轴的标签和坐标
plt.ylabel('Moment [Nmm]') #y轴的标签和坐标
plt.savefig('moment.tif',dpi = 600,bbox_inches = 'tight',pad_inches = 0.5) #图片保存，包括名称，像素。
plt.show() #图片运行显示
```
