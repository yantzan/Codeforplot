```python
#Python3.6AnacondaJupyter Notebook
#Python
import xlrd
import numpy as np
import matplotlib.pyplot as plt

#txt
data = np.loadtxt('C:/Users/12/Desktop/ts.txt')

x = data[:,0] #x
y = data[:,1] #y
fig,ax = plt.subplots() #
ax.plot(x,y) #
ax.set_title('Relationship between moment and time') #
fig.set_size_inches(16,9) #

plt.plot(x,y, ls = '--',c = 'g', label = 'momemt') #
plt.legend() #
plt.grid(True, lw = 0.5, ls = '--', color = '0') #TrueFalse

plt.xlabel('Time [s]') #x
plt.ylabel('Moment [Nmm]') #y
plt.savefig('moment.tif',dpi = 600,bbox_inches = 'tight',pad_inches = 0.5) #
plt.show() #
```
