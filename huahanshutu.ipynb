```python

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties

myfont = FontProperties(fname = 'C:/Windows/Fonts/times.ttf' )
 
t = np.arange(0.524,1.047,0.01)
 
y = 25*(57.5+61.15*np.cos(2*t)+13.98*np.cos(4*t))/(16.57-13.98*np.cos(4*t))

plt.plot(t,y,c = 'r', ls = '--',lw = 2)
plt.xlabel('ply angle / rad',fontproperties = myfont)
plt.ylabel('radius / mm',fontproperties = myfont)
plt.savefig('hanshu.tif',dpi = 300,bbox_inches = 'tight',pad_inches = 0.5)
plt.show()
```
