```python
import xlrd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

myfont = FontProperties(fname = 'C:/Windows/Fonts/arial.ttf' )
#fig, axes = plt.subplots(1, 1, figsize=(16, 9), dpi=100)

book = xlrd.open_workbook("C:/Users/12/Desktop/scatter2.xlsx")#打开Excel文件
sheetName = book.sheet_names() #取Excel文件的所有工作表，即sheet（i）
sheet1 = book.sheet_by_name(sheetName[0])#将第一个工作表取出来
sheet2 = book.sheet_by_name(sheetName[1])

data_1col0 = sheet1.col_values(0)[0:160]#取工作表1中的第一列数据，为时间
data_1col1 = sheet1.col_values(1)[0:160]#去工作表中的第三列数据，为能量

data_2col0 = sheet2.col_values(0)[0:40]#取工作表1中的第一列数据，为时间
data_2col1 = sheet2.col_values(1)[0:40]#去工作表中的第三列数据，为能量

x1 = data_1col0 #将第一列数据赋给x1
y1 = data_1col1#将第一列数据赋给y1

x2 = data_2col0 #将第一列数据赋给x2
y2 = data_2col1#将第一列数据赋给y2

plt.ylim(ymax=60)
plt.xlim(xmin=-10,xmax=70)
plt.plot(x1,y1, ls = '-',marker='o', label = 'Loading') #设置曲线的格式，包括线型，颜色和标签
plt.plot(x2,y2, ls = '-',marker='v', label = 'Unloading') #设置曲线的格式，包括线型，颜色和标签 c = 'darkorange',
plt.legend() #显示图例
#plt.grid(False, lw = 0.5,which='major' ,ls = '--', color = '0') #刻度网格的显示，如果不喜欢将True改成False

plt.xlabel('Displacement / mm',fontproperties = myfont) #x轴的标签和坐标
plt.ylabel('Load / N',fontproperties = myfont) #y轴的标签和坐标

legend = plt.legend(prop = myfont)
plt.savefig('load-dispcurve.tif',dpi = 300,bbox_inches = 'tight',pad_inches = 0.5)
plt.show()
```
