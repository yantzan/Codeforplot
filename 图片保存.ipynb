```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(0,5,100)
y = np.sin(x)
fig,ax = plt.subplots()
ax.yaxis.set_ticks_position('right')
ax.plot(x,y)
ax.set_title('sample')
fig.set_size_inches(3.6,2)
plt.savefig('1.tif',dpi = 600,bbox_inches = 'tight',pad_inches = 0.5)
plt.show()
```
